## Final Evaluation

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creates arrays for the models and tuning methods used and the results obtained
Models = ['Decision Tree', 'SVM', 'Neural Networks']
Tuning_Method = ['Grid Search', 'Randomized Search']
Training_Score = [0.673, 0.716, 0.749]
Test_Score = [0.693, 0.718, 0.719]
GS_Accuracy = [0.715, 0.721, 0.742]
RS_Accuracy = [0.721, 0.723, 0.745]
```

```python
# Creates a dual bar chart comparing the training and test scores for each model
df = pd.DataFrame({'Model': Models, 'Training Score': Training_Score, 'Test Score': Test_Score})

df = df.set_index('Model')

ax = df.plot(kind='bar', figsize=(10, 6))

plt.title('Comparison of Training and Test Scores for Different Models')
plt.xlabel('Model')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.legend(title='Score Type')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.3f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 7), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Creates a dual bar chart comparing the accuracy scores for each tuning method
df = pd.DataFrame({'Models': Models, 'Grid Search': GS_Accuracy, 'Randomized Search': RS_Accuracy})

df = df.set_index('Models')

ax = df.plot(kind='bar', figsize=(10, 6))

plt.title('Comparison of Grid Search and Randomized Search Accuracy Scores')
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.xticks(rotation=0)
plt.legend(title='Tuning Method')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.3f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 7), textcoords='offset points')

plt.tight_layout()
plt.show()
```
